# Global Climate Trends & Risk Monitor (GCT-RM)

## 1. Introducción

El proyecto **Global Climate Trends & Risk Monitor (GCT-RM)** tiene como objetivo transformar datos meteorológicos en información útil para apoyar la toma de decisiones. A través de la automatización del proceso de obtención, limpieza y análisis de datos climáticos, el sistema busca identificar tendencias, detectar anomalías y generar visualizaciones que faciliten la interpretación de la información.

## 2. Objetivos

### Objetivo General

Desarrollar una herramienta que permita monitorear tendencias climáticas mediante el análisis automatizado de datos meteorológicos.

### Objetivos Específicos

* Obtener datos climáticos desde una API confiable.
* Procesar y limpiar la información obtenida.
* Analizar tendencias y anomalías climáticas.
* Visualizar los resultados de manera clara.
* Facilitar la toma de decisiones basada en evidencia.

## 4. Obtención de datos


Para la recopilación de información meteorológica se implementó la clase `ClimaClient`, encargada de establecer la comunicación con la API **Open-Meteo**.

Este módulo automatiza el proceso de consulta de datos climáticos y realiza validaciones mediante expresiones regulares (Regex) para garantizar la calidad de la información obtenida antes de su procesamiento.

Las principales funciones del módulo son:

* Construir dinámicamente la URL de consulta hacia la API.
* Realizar la solicitud HTTP y obtener la respuesta en formato JSON.
* Validar coordenadas, fechas y registros utilizando expresiones regulares.
* Procesar la respuesta y convertirla en una lista de diccionarios con un formato uniforme.
* Entregar los datos listos para la etapa de limpieza y análisis.

### Variables obtenidas

Los registros recuperados contienen las siguientes variables:

* Fecha
* Hora
* Temperatura (°C)
* Humedad relativa (%)

* Precipitación (mm)
